# 20.2 Finite Difference Approximating Derivatives

Finite Difference Approximation - Use values of the function in the neighborhood to approximate goal.

Forward Difference - Estimates the slope of x_j using line that connects (x_j, f(x_j)) and (x_jplus1, f(x_jplus1))

f'(x_j) = (f(x_jplus1) - f(x_j)) / (x_jplus1 - x_j)


Backward Difference - Estimates the slope of x_j using line that connects (x_j, f(x_j)) and (x_jminus1, f(x_minus1))

f'(x_j) = (f(x_j) - f(x_jminus1)) / (x_j - x_jminus1)

Central Difference - Estimates the slope of x_j using line that connects (x_jminus1, f(x_minus1)) and (x_jplus1, f(x_jplus1))

O(h) = h(α + ϵ(h))
  - α is a constant
  - ϵ(h) is a function of h that goes to 0 as h -> 0
  - describes the accuracy of the formula for approximating derivatives

O(h^p)
- p is the order of the approximation. 
- the higher the value of p, the faster the equation reaches 0



Taylor Series based Approximations

Foward Difference Formula:

f'(x_j) = (f(x_jplus1) - f(x_j)) / h

Backward Difference Formula:

f'(x_j) = (f(x_j) - f(x_jminus1)) / h

Central Difference Formula: 

f'(x_j) = (f(x_jplus1) - f(x_jminus1)) / 2h


```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

h = 0.1
x = np.arange(0, 2*np.pi, h)
y = -np.sin(x) 

f_diff = np.diff(y)/h 
x_diff = x[:-1:] 
exact = -np.cos(x_diff) 

plt.figure(figsize = (12, 8))
plt.plot(x_diff, f_diff, '-', label = 'Forward Difference Approx.')
plt.plot(x_diff, exact, label = 'Exact Solution')
plt.legend()
plt.show()
```

# 20.4 Numerical Differentiation with Noise


Noise - Value is off by a small amount from what it would be if it were computed from a pure mathematical function

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

x = np.arange(0, 2*np.pi, 0.01) 

omega = 120
epsilon = 0.1

y = np.sin(x) 
y_noise = y + epsilon*np.cos(omega*x)

plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', label = 'cos(x) + noise')
plt.plot(x, y, 'b-', label = 'cos(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
